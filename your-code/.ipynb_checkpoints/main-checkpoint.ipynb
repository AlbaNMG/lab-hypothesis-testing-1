# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here:

#1.Set hypothesis

#H0: hourly workers = 30
#H1: hourly workers != 30
```

```python
#2.Significane level

#95% confidence level-> 1-0.95= 0.05 (5%)
```

```python
#3.Sample

sample=salaries[salaries['Salary or Hourly']=='Hourly']['Hourly Rate'].sample(30)
sample
```

```python
#4.Statistic/ 5.P-value

st.ttest_1samp(sample,30)
```

```python
#6.Decide

#The p-value is not lower than the significance level (5%). For that reason, I can not reject the null hipothesis (H0:hourly workers = 30)
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our hypothesis is greater than, then to reject, the test statistic must also be positive.

```python
# Your code here:

#Police Force

#1.Set hypothesis

#H0: hourly workers <= 86.000
#H1: hourly workers > 86.000
```

```python
#2.Significane level

#95% confidence level-> 1-0.95= 0.05 (5%)
```

```python
#3.Sample

#Creating a new column to be able to filter by "Hourly"-> 1 or "Salary" -> 0 based on column "Salary or Hourly"
salaries['Hourly'] = np.where(salaries['Salary or Hourly']=='Hourly',1,0)

#Sample
sample_police=salaries[(salaries['Department']=='POLICE') & (salaries['Hourly']==0)]['Annual Salary'].sample(30)
sample_police
```

```python
#4.Statistic/ 5.P-value

st.ttest_1samp(sample_police,86000,alternative='greater')
```

```python
#6.Decide

#If our hypothesis is "greater than", then to reject, the test statistic must also be positive.In this case, the statistic is positive, so I reject the H0.
```
